```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

def generate_circle(center, radius, num_points=5000):
    theta = np.linspace(0, 2*np.pi, num_points)
    x = center[0] + radius * np.cos(theta)
    y = center[1] + radius * np.sin(theta)
    return x, y

# Parameters
system_size = 2028
num_circles = 300
circle_radius = 50

# Generate random centers within the system size
centers = np.random.uniform(0, system_size, size=(num_circles, 2))

# Generate circles
circles = [generate_circle(center, circle_radius) for center in centers]

# Plot circles
plt.figure(figsize=(8, 8))
for circle in circles:
    plt.scatter(circle[0], circle[1], s=1)

plt.axis('equal')
plt.xlim(0, system_size)
plt.ylim(0, system_size)
plt.title('Monodisperse Circles')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Calculate statistics
radii = np.full(num_circles, circle_radius)
mean_radius = np.mean(radii)
variance_radius = np.var(radii)
skewness_radius = skew(radii)
std_dev_radius = np.std(radii)

# Display statistics in a table
print("Statistical Characterization of monodisperse circle Radii:")
print("----------------------------------------------")
print(f"Mean Radius: {mean_radius:.2f}")
print(f"Variance of Radius: {variance_radius:.2f}")
print(f"Skewness of Radius: {skewness_radius:.2f}")
print(f"Standard Deviation of Radius: {std_dev_radius:.2f}")

# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(radii, bins=30, color='b', edgecolor='black', alpha=0.7)
plt.xlabel('Radius')
plt.ylabel('Frequency')
plt.title('Histogram of Circle Radii')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

def generate_rod(center, length, angle):
    dx = (length / 2) * np.cos(angle)
    dy = (length / 2) * np.sin(angle)
    start_point = (center[0] - dx, center[1] - dy)
    end_point = (center[0] + dx, center[1] + dy)
    return start_point, end_point

def plot_rods(centers, lengths, angles):
    for center, length, angle in zip(centers, lengths, angles):
        start_point, end_point = generate_rod(center, length, angle)
        plt.plot([start_point[0], end_point[0]], [start_point[1], end_point[1]])

# Parameters
num_rods = 300
image_size = 2048
lengths = np.random.uniform(50, 150, size=num_rods)  # Random lengths within a range
angles = np.random.uniform(0, 2*np.pi, size=num_rods)  # Random orientations
centers = np.random.rand(num_rods, 2) * image_size

# Generate and plot rods
plot_rods(centers, lengths, angles)
plt.axis('equal')
plt.xlim(0, image_size)
plt.ylim(0, image_size)
plt.title('Rods of Finite Lengths')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Statistical characterization
mean_length = np.mean(lengths)
std_dev_length = np.std(lengths)
variance_length = np.var(lengths)
skewness_length = skew(lengths)

# Display statistics in a table
print("Statistical Characterization of Rod Lengths:")
print("--------------------------------------------")
print(f"Mean Length: {mean_length:.2f}")
print(f"Standard Deviation of Length: {std_dev_length:.2f}")
print(f"Variance of Length: {variance_length:.2f}")
print(f"Skewness of Length: {skewness_length:.2f}")

# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(lengths, bins=30, color='b', edgecolor='black', alpha=0.7)
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.title('Histogram of Rod Lengths')
plt.grid(True)
plt.show()
```

```python
#monodisperse hardcore sphere form factor
import numpy as np
import matplotlib.pyplot as plt

# Constants
density_core = 1.0
density_shell = 0.8
density_solv = 0.6
radius_core = 10  # nm
thickness_shell = 2  # nm
radius_shell = radius_core + thickness_shell

# Volume calculations
volume_core = (4/3) * np.pi * radius_core**3
volume_shell = (4/3) * np.pi * radius_shell**3 - volume_core

# q range
delta_q = 0.01
q_values = np.arange(0.01, 5.01, delta_q)

# j1 function
def j1(x):
    return (np.sin(x) - x * np.cos(x)) / x**2

# P(q) calculation
def P(q):
    term1 = 3 * volume_core * (density_core - density_shell) * j1(q * radius_core) / (q * radius_core)
    term2 = 3 * volume_shell * (density_shell - density_solv) * j1(q * radius_shell) / (q * radius_shell)
    return (1/volume_shell) * (term1 + term2)**2

# Calculate P(q)
P_values = np.array([P(q) for q in q_values])

# Plot
plt.figure(facecolor='white')

# Taking the logarithm of q and P values
log_q_values = np.log10(q_values)
log_P_values = np.log10(P_values)

plt.plot(log_q_values, log_P_values)
plt.xlabel('log(q)')
plt.ylabel('log(P)')
plt.title('Log-Log Plot of P over q')
plt.grid(True)
plt.show()

```

```python
##bidisperse particle system
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
from scipy.integrate import trapz

# Definition of the normalized Schulz distribution function
def f(R, z, R_avg):
    return ((z + 1)**(z + 1) * R**z * np.exp(- (z + 1) * R / R_avg)) / (R_avg**(z + 1) * gamma(z + 1))

# Definition of the form factor for a sphere (simplified version)
def F(x):
    return 3 * (np.sin(x) - x * np.cos(x)) / x**3

# Definition of the scattering intensity I(q)
def I(q, R_values, z, R_avg, N0, delta_rho):
    return N0 * delta_rho**2 * np.trapz(f(R_values, z, R_avg) * R_values**6 * F(q * R_values)**2, R_values)

# Parameters for polydispersity
N0 = 20e5   # Total number of particles
delta_rho = 0.1  # Difference in scattering length density

# R and q value arrays
R_values = np.linspace(0.1, 20, 400)
q_values = np.logspace(-2, 1, 400)

# Distribution parameters for the distributions
params = [
    {'z': 10, 'R_avg': 5.0, 'label': 'Distribution 1'},
    {'z': 15, 'R_avg': 18.0, 'label': 'Distribution 2'}
]

f_R_averages = []
I_q_averages = []

# Computing average f(R) and I(q) across distributions
for param in params:
    f_R_averages.append(f(R_values, param['z'], param['R_avg']))
    I_q_averages.append([I(q, R_values, param['z'], param['R_avg'], N0, delta_rho) for q in q_values])

f_R_avg = np.mean(f_R_averages, axis=0)
I_q_avg = np.mean(I_q_averages, axis=0)

# Integrating to find total distribution and scattering intensity
total_distribution = trapz(f_R_avg, R_values)
total_scattering = trapz(I_q_avg, q_values)

# Plot log of total scattering intensity vs log(q)
plt.figure(figsize=(8, 6), facecolor='white')
plt.loglog(q_values, I_q_avg, label='Total Scattering Intensity')
plt.xlabel('log(q)')
plt.ylabel('log(I(q))')
plt.title('Total Scattering Intensity vs log(q)')
plt.legend()
plt.grid(True)
plt.show()

# Plot size distribution
plt.figure(figsize=(8, 6), facecolor='white')
plt.plot(R_values, f_R_avg, label='Average Particle Size Distribution')
plt.xlabel('R')
plt.ylabel('f(R)')
plt.title('Bimodal_Schulz_distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Improved  monodisperse cylindrical rod form factor
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.special import j1

# Constants
density_cyl = 1.0  # example value for rho_cyl
density_solv = 0.9  # example value for rho_solv
r = 1.0  # radius in arbitrary units
L = 5.0  # length in arbitrary units
V_cyl = np.pi * r**2 * L

# Function f(q, alpha)
def f(q, alpha):
    H = L  # height of the cylinder is its length
    return 2 * (density_cyl - density_solv) * V_cyl * np.sinc(q * H * np.cos(alpha)) * (j1(q * r * np.sin(alpha)) / (q * r * np.sin(alpha)))

# Probability density function P(q)
def P(q):
    result, _ = quad(lambda alpha: f(q, alpha)**2 * np.sin(alpha), 0, np.pi/2)
    return result / V_cyl

# Values of q to evaluate
q_values = np.linspace(0.1, 10, 1000)
P_values = np.array([P(q) for q in q_values])

# Plotting log P over log q
plt.figure(facecolor='white')
plt.plot(np.log(q_values), np.log(P_values), label='log P(q)')
plt.xlabel('log q')
plt.ylabel('log P(q)')
plt.title('Log of P over log q')
plt.legend()
plt.grid(True)
plt.show()
```
